## Histograms

```python
# Loading Iris Dataset
iris=pd.read_csv("iris.data", header=None)
iris.columns=["sepal_length","sepal_width", "petal_length",
              "petal_width", "species"]
```

```python
iris.head()
```

```python
# alpha value provodes trasperency
iris['sepal_length'].plot.hist(alpha=0.7)
```

```python
# stacked histogram
iris.plot.hist(alpha=1, stacked=True)
```

```python
# bins parameter provides the frequency distribution scale at x-asix
iris['sepal_length'].plot.hist(alpha=1, stacked=True, bins=25)
```

```python
# horizontal histogram
iris["sepal_width"].plot.hist(orientation="horizontal")
```

```python
iris["sepal_length"].diff().hist()
```

```python
iris.hist(color="blue", alpha=1, bins=20)
```

## Boxplot charts

```python
iris.plot.box()
```

```python
# adding colors to boxplot
colors={'boxes': 'Red', 'whiskers': 'blue','medians': 'Black', 'caps': 'Green'}
```

```python
iris.plot.box(color=colors)
```

```python
# horizontal boxplot
iris.plot.box(vert=False)
```

## Scatter Plots

```python
df.plot.scatter(x='A', y='B')
```

```python
iris.plot.scatter(x='sepal_length', y='sepal_width',
                  c='petal_length', s=100)
```

```python
# adding width to scatter points
iris.plot.scatter(x='sepal_length', y='sepal_width',
                  s=iris['petal_length'] * 50)
```

## Hexagonal bin charts

```python
iris.plot.hexbin(x="sepal_length", y="sepal_width", gridsize=25)
```

```python
iris.plot.hexbin(x="sepal_length", y="sepal_width", gridsize=10)
```

## Pie Charts

```python
iris_avg=iris["petal_width"].groupby(iris["species"]).mean()
iris_avg
```

```python
iris_avg.plot.pie()
```

```python
iris_avg.plot.pie(labels=["setosa","versicolor", "virginica"],
                  colors=list("brg"), fontsize=25, figsize=(10,10))
```

```python
iris_avg_2=iris[["petal_width",
                 "petal_length"]].groupby(iris["species"]).mean()
```

```python
# creating subplot
iris_avg_2.plot.pie(subplots=True)
```
